```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 50
std_dev = 5
size = 100

data = np.random.normal(mean, std_dev, size)

mean_value = np.mean(data)
median_value = np.median(data)
std_dev_value = np.std(data)

print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Standar Deviasi: {std_dev_value:.2f}")

x = np.linspace(min(data), max(data), 1000)
pdf = norm.pdf(x, mean_value, std_dev_value)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, color='skyblue', label='Distribusi Normal')

plt.axvline(mean_value, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_value:.2f}')
plt.axvline(median_value, color='green', linestyle='--', linewidth=2, label=f'Median: {median_value:.2f}')
plt.axvline(mean_value + std_dev_value, color='orange', linestyle=':', linewidth=2, label=f'+1 Std Dev: {mean_value + std_dev_value:.2f}')
plt.axvline(mean_value - std_dev_value, color='orange', linestyle=':', linewidth=2, label=f'-1 Std Dev: {mean_value - std_dev_value:.2f}')

plt.title("Diagram Garis Distribusi Normal")
plt.xlabel("Nilai")
plt.ylabel("Probabilitas")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'x': [5, 10, 15, 20, 25],
        'y': [10, 20, 20, 30, 30]}
df = pd.DataFrame(data)

mean = df.mean()
median = df.median()
mode = df.mode().iloc[0]

print("\nStatistik:")
print(f"Mean:\n{mean}\n")
print(f"Median:\n{median}\n")
print(f"Mode:\n{mode}\n")

plt.figure(figsize=(12, 6))

plt.plot(df.index, df['x'], marker='o', color='dodgerblue', linestyle='-', linewidth=2, label='x values')
plt.plot(df.index, df['y'], marker='s', color='tomato', linestyle='-', linewidth=2, label='y values')

plt.axhline(mean['x'], color='blue', linestyle='--', label=f'x mean: {mean["x"]:.2f}')
plt.axhline(median['x'], color='blue', linestyle='-.', label=f'x median: {median["x"]:.2f}')
plt.axhline(mode['x'], color='blue', linestyle=':', label=f'x mode: {mode["x"]:.2f}')

plt.axhline(mean['y'], color='red', linestyle='--', label=f'y mean: {mean["y"]:.2f}')
plt.axhline(median['y'], color='red', linestyle='-.', label=f'y median: {median["y"]:.2f}')
plt.axhline(mode['y'], color='red', linestyle=':', label=f'y mode: {mode["y"]:.2f}')

plt.title('Diagram Garis Distribusi Kolom x dan y', fontsize=14, fontweight='bold')
plt.xlabel('Index', fontsize=12)
plt.ylabel('Nilai', fontsize=12)
plt.legend(loc='upper left', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
